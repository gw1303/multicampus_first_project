```python
# 전체장르 종류 df
genreDf = pd.DataFrame(pd.read_json('data/genre_gn_all.json', encoding='utf-8', typ='series'), columns=['genre'])
genreDfIndex = list(genreDf.index)
genreDfIndex

genreDf
```

```python
# 전체 장르 id 중에서 메인장르 id만 추출해 리스트 생성
import re 
mainGenrePat = re.compile('.{4}00')
mainGenreList = []

for i in genreDfIndex :
    if re.search(mainGenrePat, i) :
        mainGenreList.append(re.search(mainGenrePat, i).group())

mainGenreIdList = list(set(mainGenreList))
mainGenreIdList
```

```python
# 메인장르 id를 인덱스로 가진 장르명 df 추출
mainGenreDf = genreDf.loc[mainGenreIdList]

mainGenreList = mainGenreDf['genre'].tolist()
len(mainGenreList)  # 32종류
mainGenreList
```

```python
mainGenreDict = dict(zip(mainGenreIdList,[i.split('/') for i in mainGenreList]))
mainGenreDict
```

```python
import itertools
mainGenreSepList = list(itertools.chain(*list(mainGenreDict.values())))
mainGenreSepList = list(set(mainGenreSepList))
mainGenreSepList
```

```python
def findMainGenre(message):
    findGenre = list()
    import re
    for i in mainGenreSepList :
#         if i in message :
#             findGenre.append(i)

        re.compile(i)    
        if re.search(i, message) :
            findGenre.append(re.search(i, message).group())
    
    return findGenre
```
